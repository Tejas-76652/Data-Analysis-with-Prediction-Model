```python
import matplotlib.pyplot as plt

def predict_future(region='India', segment='Clients', dataframe=Tfunnel):
    # Filter the dataset based on the specified region and segment
    filtered_df = dataframe[(dataframe['Region'] == region) & (dataframe['Segment'] == segment)]

    # Sort the dataset by Year and Month in ascending order
    filtered_df = filtered_df.sort_values(['Year', 'Month'], ascending=True)

    # Select the required columns for prediction
    prediction_df = filtered_df[['Year', 'Month', 'Lv5_Visitors']].copy()

    # Set the Year and Month columns as the index
    prediction_df.set_index(['Year', 'Month'], inplace=True)

    # Convert the index to datetime format
    prediction_df.index = pd.to_datetime(prediction_df.index.map('{0[0]}-{0[1]}'.format))

    # Perform the prediction for the next 6 months using the last known value
    last_known_value = prediction_df.iloc[-1]['Lv5_Visitors']
    future_dates = pd.date_range(start=prediction_df.index[-1], periods=7, freq='M')[1:]
    future_values = [last_known_value] * len(future_dates)
    predicted_df = pd.DataFrame({'Lv5_Visitors': future_values}, index=future_dates)

    # Plotting the predicted values
    plt.figure(figsize=(10, 6))
    plt.plot(prediction_df.index, prediction_df['Lv5_Visitors'], marker='o', label='Actual')
    plt.plot(predicted_df.index, predicted_df['Lv5_Visitors'], marker='o', label='Predicted')
    plt.xlabel('Date')
    plt.ylabel('Level 5 Visitors')
    plt.title('Predicted Level 5 Visitors for {} - {}'.format(region, segment))
    plt.legend()
    plt.show()

# Calling the function with default values
predict_future()
```

**Q.** Generates the MAPE and RMSE of your prediction of the year 2022, 2021 & 2020 for the given parameters.

**Ans.**

```python
import numpy as np

def predict_future(region='India', segment='Clients', dataframe=Tfunnel):
    # Filter the dataset based on the specified region and segment
    filtered_df = dataframe[(dataframe['Region'] == region) & (dataframe['Segment'] == segment)]

    # Sort the dataset by Year and Month in ascending order
    filtered_df = filtered_df.sort_values(['Year', 'Month'], ascending=True)

    # Select the required columns for prediction
    prediction_df = filtered_df[['Year', 'Month', 'Lv5_Visitors']].copy()

    # Set the Year and Month columns as the index
    prediction_df.set_index(['Year', 'Month'], inplace=True)

    # Convert the index to datetime format
    prediction_df.index = pd.to_datetime(prediction_df.index.map('{0[0]}-{0[1]}'.format))

    # Initialize lists to store MAPE and RMSE for each year
    mape_list = []
    rmse_list = []

    # Iterate over each year
    years = prediction_df.index.year.unique()
    for year in years:
        # Filter the data for the current year
        year_df = prediction_df[prediction_df.index.year == year]

        # Perform the prediction for the next 6 months using the last known value
        last_known_value = year_df.iloc[-1]['Lv5_Visitors']
        future_dates = pd.date_range(start=year_df.index[-1], periods=7, freq='M')[1:]
        future_values = [last_known_value] * len(future_dates)
        predicted_df = pd.DataFrame({'Lv5_Visitors': future_values}, index=future_dates)

        # Calculate the MAPE and RMSE
        actual_values = year_df['Lv5_Visitors'].values
        predicted_values = np.repeat(predicted_df['Lv5_Visitors'].values[0], len(actual_values))  # Adjust length of predicted values
        mape = np.mean(np.abs((actual_values - predicted_values) / actual_values)) * 100
        rmse = np.sqrt(np.mean((actual_values - predicted_values) ** 2))

        # Append the MAPE and RMSE to the respective lists
        mape_list.append(mape)
        rmse_list.append(rmse)

        # Plotting the predicted values
        plt.figure(figsize=(10, 6))
        plt.plot(year_df.index, year_df['Lv5_Visitors'], marker='o', label='Actual')
        plt.plot(predicted_df.index, predicted_df['Lv5_Visitors'], marker='o', label='Predicted')
        plt.xlabel('Date')
        plt.ylabel('Level 5 Visitors')
        plt.title('Predicted Level 5 Visitors for {} - {} - {}'.format(region, segment, year))
        plt.legend()
        plt.show()

    # Print MAPE and RMSE for each year
    for i, year in enumerate(years):
        print("MAPE for {} - {}: {:.2f}%".format(region, year, mape_list[i]))
        print("RMSE for {} - {}: {:.2f}".format(region, year, rmse_list[i]))
        print()

# Calling the function with default values

predict_future()
```

**2.** Perform an AB testing to find which variation whether control or treatment is better.

**Ans.** To Determine which one is best , let 's take two factors:
  
  * t-test and

  * Conversion rate 

```python
from scipy.stats import ttest_ind

# Separate the control and treatment groups
control_group = ab[ab['Variations'] == 'Control']
treatment_group = ab[ab['Variations'] == 'Treatment']

# Perform the t-test
t_stat, p_value = ttest_ind(control_group['Clicks'], treatment_group['Clicks'])

# Print the results
print('Results of the A/B test:')
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```

```python
# Calculate the conversion rate for treatment group
treatment_conversion_rate = ab[ab['Variations'] == 'Treatment']['Conversion_Rate'].sum() / ab[ab['Variations'] == 'Treatment']['Visitors'].sum()

# Calculate the conversion rate for control group
control_conversion_rate = ab[ab['Variations'] == 'Control']['Conversion_Rate'].sum() / ab[ab['Variations'] == 'Control']['Visitors'].sum()

# Print the conversion rates
print("Conversion Rate for Treatment group:", treatment_conversion_rate)
print("Conversion Rate for Control group:", control_conversion_rate)
```

In this case, the conversion rate for the treatment group is 0.001484 (or 0.1484%), while the conversion rate for the control group is 0.000218 (or 0.0218%). The treatment group has a higher conversion rate.

The p-value of the A/B test is extremely low (6.006205939916319e-18), indicating strong statistical significance. This suggests that the difference in conversion rates between the treatment and control groups is unlikely to occur by chance.

Based on these results, we can conclude that the treatment group has a higher conversion rate compared to the control group, and the difference is statistically significant. Therefore, the treatment variation is likely to be better than the control.
